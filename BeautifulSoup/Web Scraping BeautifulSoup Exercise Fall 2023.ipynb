Goal: Extract job title, job link, brief description provided.

Website URL : https://pythonjobs.github.io/

```python
#Load packages
!pip install bs4
!pip install pandas
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
#get webpage html.
html = urlopen('https://pythonjobs.github.io/')
bs_html = BeautifulSoup(html, 'html.parser')
```

```python
#Inspect elements of website. Want to extract div tag of class 'job'

#Enter code below.
titles = bs_html.find_all()
#Create empty lists. Will store desired information inside them.
title = []
links = []
location = []
description = []
for item in titles:
    #Loop through all tags found from above. Find title of position

    #Enter code below.
    title1 = item.find()
    #append job title to list
    title.append(title1.get_text())
    #Find links to more detail for position.
    #Pull description.
    desc = item.find('p', {'class': 'detail'})
    description.append(desc.get_text())

    #Enter code below.
    link = item.find()
    #Append link attribute href to list. 'href' if hyperlink.
    links.append(link['href'])

    #extra. Location
    #Location, is trickier. Inspecting webpage i tag

    #Enter code below.
    loc = item.find()
    #Get text of parent html element.
    loc = loc.parent.get_text()
    #Append to list.
    location.append(loc)
```

```python
data = pd.DataFrame({
    'title': title,
    'link': links,
    'desc': description,
    'loc': location

})
display(data)
```
