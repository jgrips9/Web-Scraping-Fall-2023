```python
#This code saves the pdf to a specific location.
import json
from selenium import webdriver

appState = {
    "recentDestinations": [
        {
            "id": "Save as PDF",
            "origin": "local",
            "account": ""
        }
    ],
    "selectedDestinationId": "Save as PDF",
    "version": 2
}

profile = {'printing.print_preview_sticky_settings.appState': json.dumps(appState),
           #Change path below to where you would like pdfs to be installed. This path will be different for you
           'savefile.default_directory': 'C:\\Users\\jrg363_rs\\Downloads\\pdf_downloads'}

chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option('prefs', profile)
chrome_options.add_argument('--kiosk-printing')

#Change default downloads location to above location. Not all pdfs will be downloaded to above folder. Some are downloaded automatically when clicked. 
#Change default download location so consistent for all pdf downloads. 
#Link below has instructions on how to perform this action
#https://www.howtogeek.com/231002/how-to-change-the-chrome-download-folder-location/
```

```python
#Below is test code. For only the first page.
import os
#Below is url to download pdfs from. 
url = "https://scholar.google.com/scholar?cites=3094189527534876872"
#Go to URL
driver = webdriver.Chrome(options=chrome_options)
page = driver.get(url)
    #Identify html class object that flags pdfs. find all pdf links.
links = driver.find_elements(By.CLASS_NAME, "gs_ggsd")
for link in links:
    #Wait 3 seconds for page to load.
    time.sleep(3)
    
    #Find link.
    page = link.find_element(By.TAG_NAME, "a")
    paget = link.find_element(By.TAG_NAME, "a").text
    #If link url contains 'pdf', click on link.
    if 'PDF' in paget:
        page.click()
            #Wait 3 seconds for pdf page to load
        time.sleep(3)
        #If pdf does not download automatically when clicked, a new page will open. Download the pdf and return to original page. 
        while driver.current_url != url:
                #Print current page as pdf.
            driver.execute_script('window.print();')
                #Wait 3 seconds.
            time.sleep(3)
                #return back to starting url.
            driver.back()
    #Change current URL to the next page if applicable. 
    url = driver.current_url
#Close browser.    
driver.quit()

#Note. One of the pdf links when clicked on gives a proxy error. This results in the code sometimes taking a while. Link below is the issue. Code will continue to function once error is printed.
#[PDF] nkumbauniversity.ac.ug
#http://nuir.nkumbauniversity.ac.ug/xmlui/bitstream/handle/20.500.12383/1090/Strategic%20Leadership_%20Theory%20and%20Research%20on%20Executives%2C%20Top%20Management%20Teams%2C%20and%20Boards%20%28%20PDFDrive%20%29.pdf?sequence=1&isAllowed=y
```

```python
#Repeat process traversing multiple pages.
#Set starting page number
pagenum = 1
#Set ending page number. 
maxpages = 2
#Same process below as done in above function.
url = "https://scholar.google.com/scholar?cites=3094189527534876872"
driver = webdriver.Chrome(options=chrome_options)
page = driver.get(url)
while pagenum <= maxpages:
    links = driver.find_elements(By.CLASS_NAME, "gs_ggsd")
    for link in links:
        time.sleep(2)
    
        page = link.find_element(By.TAG_NAME, "a")
        paget = link.find_element(By.TAG_NAME, "a").text
        if 'PDF' in paget:
            page.click() 
            time.sleep(2)
            if driver.current_url != url:
                driver.execute_script('window.print();')
                time.sleep(2)
                driver.back()
    #Go to next page and record which page on so program knows when to stop. 
    nav = driver.find_element(By.ID, "gs_n")
    nextb = nav.find_elements(By.TAG_NAME, 'a')
    nextb[len(nextb) - 1].click()
    pagenum = pagenum + 1
    url = driver.current_url
driver.quit()

#Note. One of the pdf links when clicked on gives a proxy error. This results in the code sometimes taking a while. Link below is the issue. Code will continue to function once error is printed.
#[PDF] nkumbauniversity.ac.ug
#http://nuir.nkumbauniversity.ac.ug/xmlui/bitstream/handle/20.500.12383/1090/Strategic%20Leadership_%20Theory%20and%20Research%20on%20Executives%2C%20Top%20Management%20Teams%2C%20and%20Boards%20%28%20PDFDrive%20%29.pdf?sequence=1&isAllowed=y
```
